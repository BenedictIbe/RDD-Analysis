year = '2019'

```
month_rdd3_1 = month_rdd3.filter(lambda x: x[1][-4:] == year)
month_rdd3_1.take(10)
```

```
month_rdd4 = month_rdd3_1.flatMap(lambda arr: [(s,1) for s in arr[1].split(',')])
month_rdd4.take(10)
```

```
month_rdd_fil = month_rdd4.filter(lambda x: x != '')
month_rdd_fil.take(10)
```

```
month_rdd5 = month_rdd_fil.flatMap(lambda arr: [(s,1) for s in arr[0].split(' ')]).filter(lambda x: x[0] in ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'))
month_rdd5.take(100)
```

```
month_rdd7 = month_rdd5.reduceByKey(lambda x,y: x + y)
month_rdd7.take(12)
```

```
months_dict = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
              'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
```

```
month_rdd8 = month_rdd7.sortBy(lambda x: months_dict[x[0][:3]])
month_rdd8.take(12)
```
